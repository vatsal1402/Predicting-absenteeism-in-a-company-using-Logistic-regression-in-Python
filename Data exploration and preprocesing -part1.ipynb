```python
df_no_id = raw_data.drop(['ID'],axis = 1)
```

```python
df_no_age_id = df_no_id.drop(['Age'],axis = 1)
```

```python
sorted(df_no_age_id['Reason for Absence'].unique())
## Used to find the missing value from 0 to 28 which is 20
```

```python
reasons_dummies = pd.get_dummies(df_no_age_id['Reason for Absence'],drop_first = True)
```

```python
reasons_dummies
```

```python
reasons_dummies['check'] = reasons_dummies.sum(axis = 1)
## After checking we find that there are no missing values as such
```

```python
reasons_dummies
```

```python
reasons_dummies = reasons_dummies.drop(['check'],axis = 1)
```

```python
age_dummies = pd.get_dummies(df_no_id['Age'])
```

```python
## Removing the Reason for Absence column as we are going to incorporate the reason_dummies in the dataframe
df_no_id = df_no_id.drop(['Reason for Absence'],axis = 1)
```

```python
(reasons_dummies)
```

```python
## Grouping similiar reasons together
reason_group1 = reasons_dummies.iloc[:,0:14].max(axis = 1)
reason_group2 = reasons_dummies.iloc[:,14:17].max(axis = 1)
reason_group3 = reasons_dummies.iloc[:,17:20].max(axis = 1)
reason_group4 = reasons_dummies.iloc[:,20:].max(axis = 1)
```

```python
df_concat = pd.concat([df_no_id, reason_group1 , reason_group2 , reason_group3, reason_group4], axis = 1)
```

```python
df_concat
```

```python
df_concat.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'reason_1', 'reason_2', 'reason_3', 'reason_4']
```

```python
df_concat.columns = column_names
```

```python
df_concatenated = pd.concat([df_no_age_id , age_dummies],axis = 1)
```

```python
## Extracting the month value from the date column
month = []
for i in range(len(df_concat)):
    month.append(df_concat['Date'][i][3:5])
        
```

```python
## Another way of extracting the month value is by using the timestamp data type
df_concat['Date'] = pd.to_datetime(df_concat['Date'],format = '%d/%m/%Y')
```

```python
month_list = []
for i in range(len(df_concat)):
    month_list.append(df_concat['Date'][i].month)
```

```python
df_concat['Month Value'] = month_list
df_concat.head()
```

```python
##Extracting the day of the week from The Date column
df_concat['Date'][0].weekday()
```

```python
day_of_week_list = []
for i in range(len(df_concat)):
    day_of_week_list.append(df_concat['Date'][i].weekday())
```

```python
df_concat['Day of the Week'] = day_of_week_list
```

```python
df_concat = df_concat.drop(['Date'],axis = 1)
```
